## Feature Scaling using StandardScaler

Feature scaling is an important preprocessing step in machine learning. It ensures that all features have a similar scale, preventing features with larger values from dominating the model.

The `StandardScaler` standardizes the data using the formula:

\[
z = \frac{x - \mu}{\sigma}
\]

where:

- \(x\) = original value
- \(\mu\) = mean of the feature
- \(\sigma\) = standard deviation of the feature

After scaling:

- The mean of each feature becomes approximately **0**
- The standard deviation becomes approximately **1**

The `fit_transform()` method first learns the statistical properties of the data and then transforms it into standardized form.

The first five rows of the scaled dataset are displayed using `print(X_scaled[:5])`.

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

print(X_scaled[:5])
```

## Train-Test Split

The dataset is divided into training and testing sets using the `train_test_split()` function from Scikit-learn.

- **Training Set:** Used to train the machine learning model.
- **Testing Set:** Used to evaluate the model's performance on unseen data.

In this experiment:

- `test_size = 0.2` means **20% of the data is reserved for testing**.
- The remaining **80% of the data is used for training**.
- `random_state = 42` ensures reproducibility of results by using the same random split each time the code is executed.

The shapes of the training and testing datasets are displayed to verify the split.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42
)

print(X_train.shape)
print(X_test.shape)
```

## Model Training using K-Nearest Neighbors (KNN)

In this step, the K-Nearest Neighbors (KNN) algorithm is used to train the machine learning model.

KNN is a supervised learning algorithm that classifies a data point based on the majority class of its nearest neighbors.

The parameter `n_neighbors = 3` indicates that the model considers the 3 nearest neighbors for classification.

The `fit()` function trains the model using the training dataset (`X_train` and `y_train`) by learning patterns and relationships between features and target classes.

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=3)

model.fit(X_train, y_train)
```

## Model Evaluation using Accuracy Score

After training the K-Nearest Neighbors (KNN) model, its performance is evaluated using the accuracy score.

The `accuracy_score()` function measures the proportion of correctly predicted samples out of the total number of samples.

Accuracy is calculated using the formula:

\[
\text{Accuracy} = \frac{\text{Number of Correct Predictions}}{\text{Total Number of Predictions}}
\]

The predicted labels are compared with the actual labels of the test dataset to determine the model's performance.

```python
# Check accuracy
```

```python
pred = model.predict(X_test)

print(pred[:10])

# Check accuracy
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, pred)

print("Accuracy:", accuracy)
```

## Classification Report

The classification report is used to evaluate the performance of the machine learning model for each class separately.

The `classification_report()` function provides the following metrics:

- **Precision**: Measures how many predicted samples are actually correct.
- **Recall**: Measures how many actual samples were correctly identified.
- **F1-score**: Harmonic mean of precision and recall.
- **Support**: Number of samples belonging to each class.

These metrics provide a detailed evaluation of the model beyond accuracy and help assess its performance for each species in the Iris dataset.

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, pred))
```

## Confusion Matrix Visualization

A confusion matrix is used to evaluate the performance of a classification model by comparing the actual and predicted class labels.

The `confusion_matrix()` function generates a matrix showing the number of correct and incorrect predictions for each class.

A heatmap is created using Seaborn to visualize the confusion matrix more effectively.

### Components of the Confusion Matrix

- **Rows** represent the actual classes.
- **Columns** represent the predicted classes.
- **Diagonal elements** indicate correct predictions.
- **Off-diagonal elements** indicate misclassifications.

The heatmap provides a visual representation of the model's classification performance on the Iris dataset.

```python

from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, pred)

plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```
